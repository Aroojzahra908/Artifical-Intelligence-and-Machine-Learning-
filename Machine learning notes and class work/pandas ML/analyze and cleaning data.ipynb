# loc & iloc operators

# loc:
    - Select specific rows and/or columns using loc when using the row and column names.

```python
data.loc[data["Age"] > 35, "Name"]
```

Set "Pclass" = 45 in row 3: (column name pclass)

```python
data.loc[3, 'Pclass'] = 45
```

```python
data.head()
```

Delete rows where "Age" is higher than 38:

```python
for x in data.index:
  if data.loc[x, "Age"] > 38:
    data.drop(x, inplace = True) 
```

```python
data.head()
```

# iloc:
    -Select specific rows and/or columns using iloc when using the positions in the table.
    -for example: [9:15,2:5] 
    -9:15 for rows 9 include and 15 exclude
    -after comma 2:5 means colums 2 exclude and 5 include

```python
data.iloc[9:15,2:5]    
```

You can assign new values to a selection based on loc/iloc.

```python
data.iloc[1:5, 3]="write something"
```

```python
data.head()
```

# Data cleaning:

Data cleaning means fixing bad data in your data set.Bad data could be:

         -Empty cells
         -Data in wrong format
         -Wrong data
         -Duplicates


# Dropna():
By default, the dropna() method returns a new DataFrame, and will not change the original.

```python
data.dropna()       # Return a new Data Frame with no empty cells
```

The dropna(inplace = True) will NOT return a new DataFrame, but it will remove
all rows containing NULL values from the original DataFrame.

```python
data.dropna(inplace=True)          
```

```python
data.head()
```

# Fillna():
    The fillna() method allows us to replace empty cells with a value

```python
#Replace NULL values with the number 130:
data.fillna(130, inplace = True) 
```

```python
data.tail(3)
```

```python
#Replace NULL values in the "Calories" columns with the number 130:
data["Age"].fillna(13, inplace = True)
```

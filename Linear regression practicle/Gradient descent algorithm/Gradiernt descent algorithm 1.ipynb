```python
from sklearn.datasets import make_regression
import numpy as np 
```

```python
x,y= make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt 
plt.scatter(x,y)
```

# lets apply OLS (ordinary list square)

```python
from sklearn.linear_model import LinearRegression 
```

```python
reg= LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_       # slope value after training 
```

```python
reg.intercept_          #intercept 
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x), color='red')  #regression line here  
```

# Apply gradient descent 

Assume slope is constant m = 78.35

   and let assume the starting value for intercept b=0

```python
y_predict=((78.35 * x)+0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x), color ='red' , label='OLS')
plt.plot(x,y_predict,color='green', label='b=0')
plt.legend()
plt.show()
```

```python
m=78.35   #fixed
b=0
```

-2 sum of n and i=1 (yi-mxi-b)    

in this equation we put value of b and check

```python
slope_1 = -2 * np.sum(y-m*x.ravel()-b)
slope_1
```

```python
# let learning rate =0.1
lr= 0.1
step_size=slope_1*lr
step_size
```

```python
# calculate the new intercept >> here we find new value of b 
b = b-step_size
b
```

```python
y_predict1=((78.35 * x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x), color='red', label='OLS')
plt.plot(x,y_predict1, color='green', label='b={}'.format(b))
plt.plot(x,y_predict, color='pink', label='b=0')
plt.legend()
plt.show()
```

```python
# iteration 2
slope_1 = -2 * np.sum(y-m*x.ravel()-b)
slope_1
```

```python
step_size=slope_1*lr
step_size
```

```python
b = b-step_size
b
```

```python
y_predict2=((78.35 * x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x), color='red', label='OLS')
plt.plot(x,y_predict1, color='green', label='b={}'.format(b))
plt.plot(x,y_predict2, color='orange', label='b={}'.format(b))
plt.plot(x,y_predict, color='pink', label='b=0')
plt.legend()
plt.show()
```

```python
#iteration 3
slope_1 = -2 * np.sum(y-m*x.ravel()-b)
slope_1
```

```python
step_size=slope_1*lr           # minimum becaue we close 
step_size
```

```python
b = b-step_size
b
```

```python
y_predict3=((78.35 * x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x), color='red', label='OLS')
plt.plot(x,y_predict1, color='green', label='b={}'.format(b))
plt.plot(x,y_predict2, color='orange', label='b={}'.format(b))
plt.plot(x,y_predict3, color='purple', label='b={}'.format(b))
plt.plot(x,y_predict, color='pink', label='b=0')
plt.legend()
plt.show()
```

```python
b=-100
m=78.35
lr=0.1
epochs=10

for i in range(epochs):
    slope_1 = -2 * np.sum(y-m*x.ravel()-b)
    b = b - (lr * slope_1)   # b update again and again 
    
    y_predict = m * x + b    # y chng again and again 
    
    plt.plot(x,y_predict)
    
plt.scatter(x,y)
```

```python
# if learning rate so small >> its good >> bcz large step create problem 
b=-100
m=78.35
# learning rate small >> we take small steps >> take time 
epochs=100
lr= 0.01   

for i in range(epochs):
    slope_1 = -2 * np.sum(y-m*x.ravel()-b)
    
# b update again and again 
    b = b - (lr * slope_1)
    
# y chng again and again 
    y_predict = m * x + b  
    
    plt.plot(x,y_predict)
    
plt.scatter(x,y)
```
